```python
import os
import random
import matplotlib.pyplot as plt
from skimage import io, transform
from torch.utils.data import DataLoader, Dataset
```

```python
%matplotlib inline
```

```python
root_dir_bee = 'hymenoptera_data/train/bees'
root_dir_ant = 'hymenoptera_data/train/ants'

ImageFilesBees = os.listdir('hymenoptera_data/train/bees/')
ImageFilesAnts = os.listdir('hymenoptera_data/train/ants/')
imageBee = io.imread(os.path.join(root_dir_bee, ImageFilesBees[0]))
imageAnt = io.imread(os.path.join(root_dir_ant, ImageFilesAnts[0]))

print('Image for bees:')
print('\tImage name : {}'.format(ImageFilesBees[0]))
print('\tImage size : {}'.format(imageBee.shape))
print('Image for ants:')
print('\tImage name : {}'.format(ImageFilesAnts[0]))
print('\tImage size : {}'.format(imageAnt.shape))
```

```python
i = 0
for img in zip([imageBee, imageAnt], ['Bee', 'Annt']):
    ax = plt.subplot(1, 2, i+1)
    ax.axis('off')
    ax.set_title(img[1])
    plt.imshow(img[0])
    i += 1
```

```python
class ImageDataLoader(Dataset):
    def __init__(self, root_dir_bee, root_dir_ant, transform = None):
        self.root_dir_bee = root_dir_bee
        self.root_dir_ant = root_dir_ant
        self.bee_image_files = os.listdir(root_dir_bee)
        self.ant_image_files = os.listdir(root_dir_ant)
        
    def __len__(self):
        length = len(self.bee_image_files)
        return length
    
    def __getitem__(self, index):
        choice = random.random()
        if choice <= .5:
            image = io.imread(os.path.join(self.root_dir_bee, self.bee_image_files[index]))
            return image
        else:
            image = io.imread(os.path.join(self.root_dir_ant, self.ant_image_files[index]))
            return image
```

```python
image_dataset = ImageDataLoader(root_dir_bee='hymenoptera_data/train/bees/', root_dir_ant='hymenoptera_data/train/ants/')
```
